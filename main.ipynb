```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt



poison_metrics = np.load("paccs.npy")
clean_metrics = np.load("caccs.npy")



fig = sns.lineplot(x=range(len(poison_metrics[:, 0])), y=poison_metrics[:, 0], color="red", label="Poisoned")
fig = sns.lineplot(x=range(len(clean_metrics[:, 0])), y=clean_metrics[:, 0], color="blue", label="Clean")
fig.set_title("PACC and CACC over training")
fig.set_xlabel("Training Step")
fig.set_ylabel("PACC and CACC")
fig.legend()

plt.show()
```
